### outlier remove using IQR

```python
df_copy_minmax_w=df_minmax_whole
df_copy_ronust_w=df_robust_whole
df_copy_standard_w=df_standard_whole
df_copy_minmax_s=df_minmax_severe
df_copy_robust_s=df_robust_severe
df_copy_standard_s=df_standard_severe
df_copy_minmax_ns=df_minmax_non_severe
df_copy_robust_ns=df_robust_non_severe
df_copy_standard_ns=df_standard_non_severe
b=[df_copy_minmax_w,df_copy_ronust_w,df_copy_standard_w,df_copy_minmax_s,df_copy_robust_s,df_copy_standard_s,df_copy_minmax_ns,
  df_copy_robust_ns,df_copy_standard_ns]
```

```python
for item in b:
  cols=item.columns
  for i in range(1,len(cols)):
    percentile25 = item[cols[i]].quantile(0.25)
    percentile75 = item[cols[i]].quantile(0.75)
    iqr=percentile75-percentile25
    upper_limit = percentile75 + 1.5 * iqr
    lower_limit = percentile25 - 1.5 * iqr
    item[item[cols[i]] > upper_limit]
    item[item[cols[i]] < lower_limit]
    item[cols[i]]=item[(item[cols[i]]<=upper_limit)&(lower_limit<=item[cols[i]])][cols[i]]
```
