```python
plt.figure(figsize=(50,50))
sns.boxplot(data=df_new_t)
```

```python
df_new_copy= df_new_t
b=[df_new_copy]
```

```python
for item in b:
  cols=item.columns
  for i in range(1,len(cols)):
    percentile25 = item[cols[i]].quantile(0.25)
    percentile75 = item[cols[i]].quantile(0.75)
    iqr=percentile75-percentile25
    upper_limit = percentile75 + (1.5* iqr)
    lower_limit = percentile25 - (1.5* iqr)
    item[item[cols[i]] > upper_limit]
    item[item[cols[i]] < lower_limit]
    item[cols[i]]=item[(item[cols[i]]<=upper_limit)&(lower_limit<=item[cols[i]])][cols[i]]
```

```python
plt.figure(figsize=(50,50))
sns.boxplot(data=df_new_copy)
```

```python
df_norm=pd.read_csv('/content/gdrive/My Drive/new covid project work/df_w_robust_remove_outlier_main.csv')
df=df_norm.set_index("Protein IDs")
df_t=df.T
df_t
```

```python
varlist=['Condition']
df_t[varlist]= df_t[varlist].apply(lambda x: x.map({'Severe' : 1 , 'Non Severe' : 0}))
df_t[varlist].head(40)
```

```python
x=pd.read_csv('/content/gdrive/My Drive/new covid project work/df_w_robust_remove_outlier_wout_con.csv')
x_n=x.set_index("Protein IDs")
x_t=x_n.T
x_t
```

```python
Y= df_t['Condition']
Y
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_t, Y, test_size=0.3, random_state=0)
```

```python
X_train
```

```python
C = [10, 1, .1, .001]

for c in C:
    clf = LogisticRegression(penalty='l1', C=c, solver='liblinear')
    clf.fit(X_train, y_train)
    print('C:', c)
    print('Coefficient of each feature:', clf.coef_)
    print('Training accuracy:', clf.score(X_train, y_train))
    print('Test accuracy:', clf.score(X_test, y_test))
    print('')
```
