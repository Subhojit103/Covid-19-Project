```python
import statsmodels.api as sm
pvals=[]
prot=[]
y=df_new_t["target"]
for i in df_new_t.columns:
    prot.append(i)
    sm_model = sm.Logit(df_new_t["target"],df_new_t[i]).fit(disp=0)
    pvals.append(sm_model.pvalues[i])
    
```

```python
sm_model.summary()
```

```python
df_new_t
```

```python
sis=pd.DataFrame()
sis["prot"]=prot
sis["pvals"]=pvals
```

```python
sis
```

```python
sum(sis["pvals"]<0.05)
```

```python
sis=sis[sis["pvals"]<0.05]
```

```python
sis_data=pd.DataFrame()
for i in sis["prot"]:
    sis_data[i]=df_new_t[i]
```

```python
sis_data["target"]=[0]*23+[1]*14
sis_data.to_csv("/content/gdrive/My Drive/sis.csv")
```
